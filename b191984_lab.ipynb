```python
import pandas as pd
```

```python
df=pd.read_csv ("MELBOURNE_HOUSE_PRICES_LESS.csv")
```

```python
#no.of rows and col
print(df.shape)
```

```python
#drop not available data
df=df.dropna()
```

```python
print(df.shape)
```

```python
rooms=df["Rooms"]
```

```python
price=df["Price"]
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
plt.hist(x=price,edgecolor="purple",bins=100)
plt.xlabel("price range")
plt.ylabel("no.of houses")
plt.title("HISTOGRAM OF PRICE")
plt.show()
```

```python
rooms_count
```
